```python
import pandas as pd
import seaborn as sns
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Загрузка данных
url = 'https://raw.githubusercontent.com/1MedvedevV/M_AI/refs/heads/main/usa_mercedes_benz_prices.csv'
df = pd.read_csv(url)

# Удаление пустых или некорректных данных
df = df[df['Price'] != 'Not Priced']
df = df[df['Mileage'] != 'Not Available']

# Преобразование числовых столбцов
df['Price'] = df['Price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df['Mileage'] = df['Mileage'].replace({' mi.': '', ',': ''}, regex=True).astype(float)
df['Rating'] = df['Rating'].astype(float)
df['Review Count'] = df['Review Count'].replace({',': ''}, regex=True).astype(float)

# Удаление строк с пропусками
df = df.dropna(subset=['Rating', 'Review Count'])

# Преобразование года выпуска
df['Year'] = df['Name'].str.extract('(\d{4})').astype(int)
df['Name'] = df['Name'].str.replace('\d{4}', '').str.strip()

# ГРАФИКИ
plt.figure(figsize=(12, 8))

# Цена от года
plt.subplot(2, 2, 1)
sns.scatterplot(x='Year', y='Price', data=df)
plt.title('Price vs Year')

# Цена от пробега
plt.subplot(2, 2, 2)
sns.scatterplot(x='Mileage', y='Price', data=df)
plt.title('Price vs Mileage')

# Цена от рейтинга
plt.subplot(2, 2, 3)
sns.scatterplot(x='Rating', y='Price', data=df)
plt.title('Price vs Rating')

# Цена от количества отзывов
plt.subplot(2, 2, 4)
sns.scatterplot(x='Review Count', y='Price', data=df)
plt.title('Price vs Review Count')

plt.tight_layout()
plt.show()

#Карта корреляции
plt.figure(figsize=(10, 8))
correlation_matrix = df[['Price', 'Mileage', 'Rating', 'Review Count', 'Year']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Подготовка зависимых и независимых переменных
X = df[['Mileage', 'Rating', 'Review Count', 'Year']]
y = df['Price']

# Логарифмическое преобразование
y = np.log1p(y)

# Нормализация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Преобразование данных в тензоры
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).view(-1, 1)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).view(-1, 1)

# Определение многослойного перцептрона (MLP)
class MLP(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, output_size):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size1)
        self.relu1 = nn.LeakyReLU()
        self.fc2 = nn.Linear(hidden_size1, hidden_size2)
        self.relu2 = nn.LeakyReLU()
        self.fc3 = nn.Linear(hidden_size2, output_size)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu1(out)
        out = self.fc2(out)
        out = self.relu2(out)
        out = self.fc3(out)
        return out

# Инициализация модели
input_size = X_train.shape[1]
hidden_size1 = 128
hidden_size2 = 64
output_size = 1

model = MLP(input_size, hidden_size1, hidden_size2, output_size)

# Оптимизатор и функция потерь
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Обучение модели
num_epochs = 400
train_losses = []
test_losses = []

for epoch in range(num_epochs):
    # Прямой проход (forward pass)
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)

    # Обратный проход и оптимизация (backward pass)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Запись ошибки на обучающей выборке
    train_losses.append(loss.item())

    # Оценка на тестовой выборке
    with torch.no_grad():
        test_outputs = model(X_test_tensor)
        test_loss = criterion(test_outputs, y_test_tensor)
        test_losses.append(test_loss.item())

    # Вывод промежуточных результатов
    if (epoch+1) % 50 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, Test Loss: {test_loss.item():.4f}')

# Оценка модели на тестовой выборке
with torch.no_grad():
    predicted = model(X_test_tensor)
    mse = criterion(predicted, y_test_tensor).item()
    accuracy = 100 - (mse / torch.mean(y_test_tensor).item() * 100)
    print(f'Точность модели: {accuracy:.2f}%')

# Построение графика ошибки
plt.figure(figsize=(10, 5))

# График ошибки на обучающей и тестовой выборке
plt.plot(train_losses, label='Training Loss')
plt.plot(test_losses, label='Test Loss')
plt.title('График ошибки на обучении и тестировании')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
